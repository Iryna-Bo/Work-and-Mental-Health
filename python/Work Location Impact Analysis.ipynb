```python
# Let's investigate whether the Work location has an impact on the following indicators: 
# work-life balance rating, stress level, productivity change, social isolation, physical activity, and sleep quality. 
# We'll use pd.crosstab and sns.heatmap for this. 
# Heatmap is a very useful tool for visualizing contingency tables, enhancing the understanding of relationships between variables.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency

data = pd.read_excel('excel/Impact_of_Remote_Work_on_Mental_Health.xlsx')
```

```python
# 1. Work Life Balance Rating:

plt.figure(figsize=(4, 4))
sns.heatmap(crosstab, annot=True, fmt="d", cmap="YlGnBu", cbar=True)
plt.title("Heatmap of Work Location vs Work Life Balance Rating")
plt.xlabel("Work_Life_Balance_Rating")
plt.ylabel("Work_Location")
plt.show()

crosstab = pd.crosstab(data['Work_Location'], data['Work_Life_Balance_Rating'])
chi2, p, dof, expected = chi2_contingency(crosstab)
print("p-value:", p)
alpha = 0.05  # Significance level
if p < alpha:
    print("The results indicate a statistically significant association between Work Location and Work-Life Balance Rating.")
else:
    print("There is no statistically significant association between Work Location and Work-Life Balance Rating.")
```

```python
# Despite the fact that the chi2_contingency test did not reveal a statistically significant relationship between the variables 
# 'Work Location' and 'Work-Life Balance,' the heatmap indicates that the largest number of surveyed employees reported the lowest level 
# of Work-Life Balance ('1') in the Remote Work segment, with 368 employees.
```

```python
# 2. Stress Level:

plt.figure(figsize=(4, 4))
sns.heatmap(crosstab, annot=True, fmt="d", cmap="YlGnBu", cbar=True)
plt.title("Heatmap of Work Location vs Stress Level")
plt.xlabel("Stress_Level")
plt.ylabel("Work_Location")
plt.show()

crosstab = pd.crosstab(data['Work_Location'], data['Stress_Level'])
chi2, p, dof, expected = chi2_contingency(crosstab)
print("p-value:", p)
alpha = 0.05  # Significance level
if p < alpha:
    print("The results indicate a statistically significant association between Work Location and Stress Level.")
else:
    print("There is no statistically significant association between Work Location and Stress Level.")
```

```python
# Although the chi2_contingency test indicates no statistically significant association between Work Location and Stress Level,
# the highest number of employees reporting High Stress levels is observed among Remote employees (590). 
# Similarly, the highest number of employees reporting Low and Medium Stress levels is observed among Onsite employees (555) 
# and Remote employees (577), respectively. However, these differences are not statistically significant.
```

```python
# 3. Productivity Change:

plt.figure(figsize=(3, 3))
sns.heatmap(crosstab, annot=True, fmt="d", cmap="YlGnBu", cbar=True)
plt.title("Heatmap of Work Location vs Productivity Change")
plt.xlabel("Productivity_Change")
plt.ylabel("Work_Location")
plt.show()

crosstab = pd.crosstab(data['Work_Location'], data['Productivity_Change'])
chi2, p, dof, expected = chi2_contingency(crosstab)
print("p-value:", p)
alpha = 0.05  # Significance level
if p < alpha:
    print("The results indicate a statistically significant association between Work Location and Productivity Change.")
else:
    print("There is no statistically significant association between Work Location and Productivity Change.")
```

```python
# Despite the fact that the chi2_contingency test did not reveal a statistically significant relationship between the variables 
# 'Work Location' and 'Productivity Change', the heatmap shows that the largest number of surveyed employees reported a decrease in productivity 
# in the Hybrid Work segment, with 591 employees. A close second is the Remote Work segment, with 588 employees.
```

```python
# 4. Social Isolation:

plt.figure(figsize=(3, 3))
sns.heatmap(crosstab, annot=True, fmt="d", cmap="YlGnBu", cbar=True)
plt.title("Heatmap of Work Location vs Social Isolation Rating")
plt.xlabel("Social_Isolation_Rating")
plt.ylabel("Work_Location")
plt.show()

crosstab = pd.crosstab(data['Work_Location'], data['Social_Isolation_Rating'])
chi2, p, dof, expected = chi2_contingency(crosstab)
print("p-value:", p)
alpha = 0.05  # Significance level
if p < alpha:
    print("The results indicate a statistically significant association between Work Location and Social Isolation Rating.")
else:
    print("There is no statistically significant association between Work Location and Social Isolation Rating.")
```

```python
# 5. Physical Activity:

# Change NaN to "None"
data['Physical_Activity'] = data['Physical_Activity'].fillna('None')

plt.figure(figsize=(4, 4))
sns.heatmap(crosstab, annot=True, fmt="d", cmap="YlGnBu", cbar=True)
plt.title("Heatmap of Work Location vs Physical Activity")
plt.xlabel("Physical_Activity")
plt.ylabel("Work_Location")
plt.show()

crosstab = pd.crosstab(data['Work_Location'], data['Physical_Activity'])
chi2, p, dof, expected = chi2_contingency(crosstab)
print("p-value:", p)
alpha = 0.05  # Significance level
if p < alpha:
    print("The results indicate a statistically significant association between Work Location and Physical Activity.")
else:
    print("There is no statistically significant association between Work Location and Physical Activity.")
```

```python
# Despite the fact that the chi2_contingency test did not reveal a statistically significant relationship between 'Work Location' and 'Physical Activity',
# the heatmap shows that the highest number of employees reporting Daily Physical Activity is in the Remote Work segment, with 586 employees. 
```

```python
# 6. Sleep Quality:

plt.figure(figsize=(3, 3))
sns.heatmap(crosstab, annot=True, fmt="d", cmap="YlGnBu", cbar=True)
plt.title("Heatmap of Work Location vs Sleep Quality")
plt.xlabel("Sleep_Quality")
plt.ylabel("Work_Location")
plt.show()

crosstab = pd.crosstab(data['Work_Location'], data['Sleep_Quality'])
chi2, p, dof, expected = chi2_contingency(crosstab)
print("p-value:", p)
alpha = 0.05  # Significance level
if p < alpha:
    print("The results indicate a statistically significant association between Work Location and Sleep Quality.")
else:
    print("There is no statistically significant association between Work Location and Sleep Quality.")
```

```python
# General conclusion: Our analysis did not confirm a significant correlation between Work Location and the six indicators mentioned above. 
# However, some notable trends can be observed in the heatmap. For instance, in the Remote Work segment, the largest number of employees reported 
# the lowest level of Work-Life Balance ('1') and the highest number of employees engaging in daily physical activity. Similarly, 
# the Hybrid Work segment showed a higher number of employees reporting a decrease in productivity, although these observations 
# were not statistically significant. These trends highlight patterns that, while not statistically conclusive, may be of interest 
# for further investigation.
```
